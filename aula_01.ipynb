# Instrução Break

```
for letra in 'Python':
  if letra == 'o':
    break
  print('Letra:' , letra)

print('\n\nContinuando o programa')
```

# Instrução Continue

```
for letra in 'Python':
  if letra == 'o':
    continue
  print('Letra:', letra)
```

# While com a instrução ELSE

```
i = 0
while i < 10:
  print(i)
  i += 1
else:
  print(f'Número é maior ou igual a 10')

print('\n\nContinuando o programa...')
```

```
i = 0
while i < 10:
  if i == 5:
    break
  print(i)
  i += 1
else:
  print(f'Número é maior ou igual a 10')

print('\n\nContinuando o programa...')
```

# Metodo writelines()

```
lista = [
         'Manipular arquivos no Python é fácil',
         'Segunda linha!',
         'Mais linha'
]
```

```
file2 = open('drive/My Drive/resultados.txt','w')
```
